```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt

def read_tensorboard_event_file(event_file):
    event_acc = EventAccumulator(event_file)
    event_acc.Reload()

    data = {'step': [], 'value': []}
    for tag in event_acc.Tags()['scalars']:
        events = event_acc.Scalars(tag)
        for event in events:
            data['step'].append(event.step)
            data['value'].append(event.value)
    return data

def plot_tensorboard_data(data):
    plt.plot(data['step'], data['value'])
    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.title('TensorBoard Data')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    event_file = '/home/s6amalia/models/snips/mt5-1B-snips/events.out.tfevents.1711728799.node-06.bender.2014183.0'  # Update with the path to your event file
    tag = 'train_loss'
    tensorboard_data = read_tensorboard_event_file(event_file)
    plot_tensorboard_data(tensorboard_data)
```
