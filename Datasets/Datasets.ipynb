```python
from datasets import Dataset
def data_loader(name , languages, split):
    values_to_drop=list(set(['en', 'de', 'es', 'fr', 'hi', 'tr'])-set(languages))
    if split == 'train':
        test=pd.read_pickle(f'/train_datasets/{name}_train.dataset')
    elif split == 'validation':
        test = pd.read_pickle(f'/train_datasets/{name}_val.dataset')
    if name == 'snips':
        ds = load_dataset('benayas/snips')
    else:
        ds=load_dataset(name, 'en')
    # Drop rows where column 'B' equals the specified values
    filtered_df = test[~test['language'].isin(values_to_drop)]
    df_ready=filtered_df.drop(columns=['language'])
    ds_s = Dataset.from_pandas(df_ready, features=ds['train'].features, preserve_index=False)
    return ds_s
```

## TTR (Type-Token Ratio) Calculation Using different methods

```python
import pandas as pd
from collections import Counter
langs=['en','de','es','fr','hi','tr']
def calculate_ttr(text):
    words = text.split()
    types = Counter(words)
    ttr = len(types) / len(words)
    return ttr
for lan in langs:
    ds_orig2 = load_dataset('xnli',lan)
    df = pd.DataFrame(ds_orig2['train'])

    df['TTR'] = df['premise'].apply(calculate_ttr)
    print(f"TTR {lan} : {df['TTR'].mean()}")
```

```python
from polyglot.text import Text
import numpy as np
def calculate_average_morphemes(text):
    try:
        polyglot_text = Text(text,hint_language_code='en' )
        total_words = len(polyglot_text.words)
        total_morphemes = sum(len(word.morphemes) for word in polyglot_text.words)
        ratio = total_morphemes / total_words
    except: ratio = np.nan
    return ratio

# English text
ds_orig2 = load_dataset('xnli','en')
df = pd.DataFrame(ds_orig2['test'])
df['TTR'] = df['premise'].apply(calculate_average_morphemes)
print(f"TTR en : {df['TTR'].mean()}")
```

```python
import nltk
from datasets import load_dataset, load_from_disk
# from nltk.tokenize import word_tokenize
# from nltk.stem import SnowballStemmer
import pandas as pd

def calculate_average_morphemes(text, language):
    # Initialize stemmer
    stemmer = nltk.stem.SnowballStemmer(language)
    # Tokenize the text
    tokens = nltk.tokenize.word_tokenize(text)
    # Count the total number of words
    total_words = len(tokens)
    # Count the total number of morphemes
    total_morphemes = sum(len(word) - len(stemmer.stem(word)) + 1 for word in tokens)
    # Calculate the average number of morphemes per word
    average_morphemes = total_morphemes / total_words
    return average_morphemes

# Create a DataFrame
if __name__ == "__main__":
    lan='es'
    if lan == 'fr':
        ds_orig2 = load_dataset('benayas/snips')
    else:
        ds_orig2 = load_from_disk("/snips_dataset/snips_" + lan + ".hf")
    # ds_orig2 = load_dataset('xnli','tr')
    df = pd.DataFrame(ds_orig2['test'])
    # Apply the function to the 'text' column
    df['average_morphemes'] = df.apply(lambda row: calculate_average_morphemes(row['text'], 'fre'), axis=1)
    print(f"TTR en : {df['average_morphemes'].mean()}")
```
